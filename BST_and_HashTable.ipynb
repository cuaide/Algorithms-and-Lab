```python
class TreeNode:
    def __init__(self, value):
        self.key = value
        self.left = None
        self.right = None
        self.parent = None
class BST:
    def __init__(self):
        self.root = None
```

```python
def insert_BST(T,z):
    x = T.root
    y = None

    while x != None:
        y = x
        if z.key < x.key:
            x = x.left
        else:
            x = x.right
    z.parent = y

    if y == None:
        T.root = z
    elif z.key < y.key:
        y.left = z
    else:
        y.right = z
```

```python
def print_BST(root, level=0, prefix="Root: "):
    if root is not None:
        print(" " * (level * 4) + prefix + str(root.key))
        if root.left is not None or root.right is not None:
            if root.left:
                print_BST(root.left, level + 1, "L--- ")
            if root.right:
                print_BST(root.right, level + 1, "R--- ")
```

```python
def transplant(T,u,v):
  if u.parent == None:
    T.root = v
  elif u == u.parent.left:
    u.parent.left = v
  else:
    u.parent.right = v
  if v != None:
    v.parent = u.parent
```

```python
def minimum(z):
  while z.left != None:
    z = z.left
  return z
```

```python
def search(root, key):
    x = root
    while x is not None and key != x.key:
        if key < x.key:
            x = x.left
        else:
            x = x.right
    return x
```

```python
def deletion_BST(T, key):
    z = search(T.root, key)
    if z is None:
        return False
    if z.left is None:
        transplant(T, z, z.right)
    elif z.right is None:
        transplant(T, z, z.left)
    else:
        y = minimum(z.right)
        if y.parent != z:
            transplant(T, y, y.right)
            y.right = z.right
            if y.right is not None:
                y.right.parent = y
        transplant(T, z, y)
        y.left = z.left
        if y.left is not None:
            y.left.parent = y
    return True
```

```python
T = BST()
for i in [23,15,10,41,20,16,35,28,17,14,39,30,3,7,12,38,26,19,47,21]:
  insert_BST(T,TreeNode(i))
```

Consider a hash table with a size of 7 and a hash function ð’‰(ð’Œð’†ð’š) = ð’Œð’†ð’š%ðŸ•.
The hash table uses chaining to handle collisions.
 Perform the following operations: â€¢ Initially insert the following keys into the hash table: 10, 22, 31, 4, 15, 28, 17, 88, 59.
  â€¢ Display the initial hash table.
  â€¢ Remove the key 17 from the hash table. (Hint: should implement the logic for finding and removing a key from the linked list associated with a specific hash value.)
  â€¢ Remove the key 21. (Hint: If a key is not present, the program should display a message indicating that the key is not found.
  â€¢ Display the hash table after the deletion.

```python
class HashTable:
  def __init__(self,size):
    self.size = size
    self.table = [None] * size

  def hash_function(self,key):
    return key % self.size

  def insert(self,key):
    index = self.hash_function(key)
    if self.table[index] is None:
      self.table[index] = [key]
    else:
      for i, k in enumerate(self.table[index]):
        if k == key:
          self.table[index][i] = key
          return
      self.table[index].append(key)

  def search(self,key):
    index = self.hash_function(key)
    if self.table[index] is None:
      return None
    else:
      for k in self.table[index]:
        if k == key:
          return k
    return None

  def delete(self,key):
    cnt = 0
    index = self.hash_function(key)
    if self.table[index] is not None:
      for i, k in enumerate(self.table[index]):
        if k == key:
          del self.table[index][i]
          cnt = 1
          return
      if cnt == 0 :
        print("Key is not found")
        return None

  def display(self):
    for index in enumerate(self.table):
      print(f"Index {index}")
```

```python
hash_table = HashTable(7)
hash_table.insert(10)
hash_table.insert(22)
hash_table.insert(31)
hash_table.insert(4)
hash_table.insert(15)
hash_table.insert(28)
hash_table.insert(17)
hash_table.insert(88)
hash_table.insert(59)
hash_table.display()
```

```python
hash_table.delete(17)
hash_table.display()
```
